```python
j = 3
filename = "ImagesOriginales/%s.jpg" %j
img = np.array(plt.imread(filename)/255, dtype="float64")
(m,n) = np.shape(img)
h = blurringKernel(m,n,pm.kernelVariance)
y = forwardOperator(img,h)
y = generateNoise(img = y, mode = 'gaussian', seed = pm.seed, var = 0.02)
```

```python
globalOptLambda = runGoldenSection(img = img, y = y, lmbdaCentroid = 0.1, blurOperator=h, alpha = [0.1,0.05,0.015]) 
```

```python

xk = y
xstar, lmbda, cost, err, xkSeq,iter = GradientDescent(x0 = xk, lmbda = globalOptLambda, blurOperator = h, maxiter = 3000, y = y, img = img, early_stopping = False)
plotResults(img,y,xstar,cost, err, globalOptLambda, iter)
```

# ora mi ricavo le diverse xkSeq 

```python

xk = y
xk, lmbda, cost, err, xkSeq, iter = GradientDescent(x0 = xk, lmbda = globalOptLambda, maxiter = 500, y = y, img = img, blurOperator = h, early_stopping = False, stopping_threshold = 1e-6)
cost[iter:] = cost[iter]
err[iter:] = err[iter]
xkSeq[iter:] = xkSeq[iter]
plotResults(img,y,xk,cost, err, globalOptLambda, iter)


xk = y
xk, lmbda, FISTA_cost, FISTA_err, FISTA_xkSeq, iter = FISTA(x0 = xk, lmbda = globalOptLambda, maxiter = 500, y = y, img = img, blurOperator = h, early_stopping = False, stopping_threshold = 1e-6)
FISTA_cost[iter:] = FISTA_cost[iter]
FISTA_err[iter:] = FISTA_err[iter]
FISTA_xkSeq[iter:] = FISTA_xkSeq[iter]
plotResults(img,y,xk,FISTA_cost, FISTA_err, globalOptLambda, iter)
```

```python
bound = np.zeros(shape = 500, dtype = 'float')
epsilon = pm.eps
lmbda = globalOptLambda
hF = np.real(np.fft.fft2(np.fft.fftshift(h)))
tau = 1.9 / (np.max(hF) + np.max(lmbda) * 8 / epsilon)

c = (np.linalg.norm(y - xstar))**2 / (2 * tau) 

for i in range(1,500):
    bound[i] = c / i
bound[0] = bound[1]

```
