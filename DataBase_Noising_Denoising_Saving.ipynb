```python
from __future__ import division
from nt_toolbox.general import *
from nt_toolbox.signal import *
%pylab inline
%matplotlib inline
%load_ext autoreload
%autoreload 2
import cv2
import skimage
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
import operator
import pandas as pd
import time
```

```python
def plotnoise_gauss(img, mode,seed,clip,var):
    if mode is not None:
        gimg = skimage.util.random_noise(img, mode=mode,seed=seed,clip=clip,var=var)
        return gimg
```

```python
col1=[]
col2=[]
col3=[]
l = dict()
s = dict()
for j in range(0,25):
    filename = "ImagesOriginales/%s.jpg" %j
    img = imread(filename)/255.0
    y=plotnoise_gauss(img,"gaussian",43,False,0.02)
    y=y*255
    path= "databaseGB/"+  str(j) + ".jpeg"
    cv2.imwrite(path,y)
    y=y/255
    epsilon = 0.4*1e-2
    for i in range(1,51):
        Lambda=i/100
        tau = 1.9 / (1 + Lambda * 8 / epsilon)
        fTV = y
        n1=y.shape[0]
        n2=y.shape[1]
        for m in range(300) :  
            repeat3 = lambda x,k: resize( repeat( x, k, axis=1), [n1,n2,k])
            Gr = grad(fTV)
            d = sqrt(epsilon**2 + sum(Gr**2, axis=2))
            G = -div(Gr / repeat3(d,2) )
            tv = sum(d.flatten())
            e = fTV -y
            fTV = fTV - tau*(e + Lambda*G)
        l[Lambda]= fTV
        s[Lambda]= str( round(snr(img, fTV), 2) ) 
    s = {k: float(v) if isinstance(v, str)  else v for k, v in s.items()}
    LambdaOpt=max(s.items(), key=operator.itemgetter(1))[0]
    col1.append(path)
    col2.append(LambdaOpt)
    col3.append("gaussian")
    
    
```
